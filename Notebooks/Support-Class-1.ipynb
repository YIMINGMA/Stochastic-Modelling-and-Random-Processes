## 2. Linear Algebra

Consider matrix $
M = \begin{pmatrix}
2 & 5 & 7 \\
6 & 1 & 8 \\
4 & 3 & 2
\end{pmatrix} \in \mathbb{R}^{3 \times 3}
$

```python
M=np.array([[2,5,7], [6,1,8], [4,3,2]]) #define a matrix 

print("shape = ", np.shape(M)) # we can return the shape of M

print("Indexing is from zero, M[0,0] =", M[0,0])

print("3rd column (M[:, 2]) = ", M[:, 2])
```

```python
print("Element wise multiplication \n", M * M)

print("\n Matrix multiplication \n", np.dot(M, M))

print("\n Powers of matrix (M^5) \n", np.linalg.matrix_power(M, 5))

print("\n Matrix Exponential (exp(M)) \n", lg.expm(M))
```

```python
print("M = \n", M)

print("Transpose \n", M.T)

print("\n Determinant \n", np.linalg.det(M))

print("\n Trace \n", np.trace(M))

print("\n Inverse \n", np.linalg.inv(M))
```

```python
evals, evecs = np.linalg.eig(M)

print("Eigenvalues \n", evals)

print("\n Corresponding eigenvectors \n", evecs)
```

## 3. Random Walks

### 3.1 Simple Random Walk

**Simple Random Walk**: Let $X_i \in \{-1,1\}, \; i \in \mathbb{N}$ be i.i.d. random variables with $\mathbb{P}[X_i = 1] = p, \mathbb{P}[X_i = -1] = q = 1-p$. Then the sequence $Y_0, Y_1,...$ defined as $Y_0 = 0$ and $Y_n = \sum_{k=1}^{n} X_k$ is the simple random walk on $\mathbb{Z}$.

```python
p = 0.1
tmax = 100
Xs = np.random.RandomState(1234).rand(tmax)
Xs[Xs < p] = 1
Xs[Xs != 1] = -1
Ys = np.cumsum(Xs)

plt.figure(figsize=(12,5))
plt.plot(range(tmax), Ys)
plt.xlabel(r"$t$", fontsize=16)
plt.ylabel(r"$Y_t$", fontsize=16)
plt.title(r"Simple Random Walk with $p = 0.1$ in $100$ steps", fontsize=18)
plt.tight_layout();
```

Recall

$\mathbb{E}[Y_n] = \mathbb{E}\left[ \sum_{k=1}^{n} X_k \right] = \sum_{k=1}^{n} \mathbb{E} [X_k]= n(2p-1)$ 

and 

$ \mathrm{Var}[Y_n] = \mathrm{Var}\left[ \sum_{k=1}^{n} x_k \right] = \sum_{k=1}^{n} \mathrm{Var}[x_k] = 4np(1-p)$

We will run $N$ simulations of a random walk and calculate numerically the average and deviations over the replications and compare this to the theoretical solution.

```python
def SRW(p, tmax, N):
    Xs = np.random.RandomState(1234).rand(N, tmax)
    Xs[Xs < p] = 1
    Xs[Xs != 1] = -1
    Xs[:, 0] = 0 # Set the initial position 0.
    Ys = np.cumsum(Xs, axis=1)
    return Xs, Ys
```

```python
tmax = 100
N = 20
p = 0.1
S = SRW(p, tmax, N)
```

#### Plot of the First Experiment

```python
# Plot the first experiment 
plt.figure(figsize=(12, 5))
plt.plot(range(tmax), S[1][0, :])
plt.xlabel(r"$t$", fontsize = 16)
plt.ylabel(r"$Y_t$", fontsize=16)
plt.title("The First Experiment", fontsize=18)
plt.tight_layout()
```

#### Plot of All Experiments

```python
plt.figure(figsize = (12,8))

for k in range(N):
    plt.plot(range(tmax), S[1][k,:], 'o', ms = 6, label ='Simulation Run: {}'.format(k)) 
    
plt.xlabel(r"$t$", fontsize=16)
plt.ylabel(r"$Y_t$", fontsize=16)
plt.legend()
plt.title("All %d Experiments"%N, fontsize=18)
plt.tight_layout();
```

#### Sample Mean Distribution

```python
Y_means = np.mean(S[1], axis=0)
Y_stds = np.std(S[1], axis = 0)
theoretical_averages = np.array(range(tmax)) * (2*p - 1)
theoretical_stds = np.sqrt(4 * np.array(range(tmax)) * p * (1-p))
```

```python
plt.figure(figsize = (12,5))
plt.errorbar(range(tmax), theoretical_averages, yerr=theoretical_stds, color='k', 
             alpha=0.8, label="Theoretical mean with std")
plt.errorbar(range(tmax), Y_means, yerr=Y_stds, ecolor='green', linewidth = 2, 
             label="Empirical mean with std")
plt.title("Empirical Average over {} realisations of SRW with error bars".format(N))
plt.xlabel(r"$t$", fontsize=16)
plt.ylabel(r"$\bar{Y}$", fontsize=16)
plt.legend(loc="lower left", fontsize=18)
plt.tight_layout();
```

#### Empirical Distribution at $n = 10$

It is possible to fit a Gaussian distribution to the histogram, using [scipy.stats.norm](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html).

```python
tmax = 100
N = 5000
p = 0.1
S = SRW(p, tmax, N)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(S[1][:, 9])
plt.title(r"Empirical Distribution at time step $n = 10$", fontsize=18)

plt.subplot(1,2,2)
plt.title(r"Empirical Distribution at time step $n = 100$", fontsize=18)
plt.hist(S[1][:, 99])

plt.tight_layout();
```
